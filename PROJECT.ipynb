```python
class Hotel:
    def __init__(self):
        self.rooms = {
            '101': {'occupant': None, 'availability': True},
            '102': {'occupant': None, 'availability': True},
            '103': {'occupant': None, 'availability': True},
    
        }

    def display_menu(self):
        print("Hotel Reservation System")
        print("1. Check Room Availability")
        print("2. Book a Room")
        print("3. Generate Reservation Confirmation")
        print("4. Exit")

    def check_availability(self):
        available_rooms = [room for room, details in self.rooms.items() if details['availability']]
        if available_rooms:
            print("Available rooms:", ', '.join(available_rooms))
        else:
            print("No rooms available.")

    def book_room(self):
        room_number = input("Enter room number: ")
        if room_number in self.rooms:
            if self.rooms[room_number]['availability']:
                guest_name = input("Enter guest name: ")
                self.rooms[room_number]['occupant'] = guest_name
                self.rooms[room_number]['availability'] = False
                print(f"Room {room_number} booked for {guest_name}.")
            else:
                print(f"Sorry, room {room_number} is already occupied.")
        else:
            print(f"Room {room_number} does not exist.")

    def generate_confirmation(self):
        room_number = input("Enter room number: ")
        if room_number in self.rooms and not self.rooms[room_number]['availability']:
            occupant = self.rooms[room_number]['occupant']
            confirmation = f"Reservation Confirmation:\nRoom Number: {room_number}\nGuest Name: {occupant}"
            print(confirmation)
        else:
            print("Invalid room number or room not occupied.")



hotel = Hotel()

while True:
    hotel.display_menu()
    choice = input("Enter your choice: ")

    if choice == '1':
        hotel.check_availability()
    elif choice == '2':
        hotel.book_room()
    elif choice == '3':
        hotel.generate_confirmation()
    elif choice == '4':
        break
    else:
        print("Invalid choice. Please try again.")
```
